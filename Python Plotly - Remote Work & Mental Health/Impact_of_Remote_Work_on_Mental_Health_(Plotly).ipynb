# Virtual meetings by Work Location

```python
plot = pgo.Figure(data=[pgo.Scatter(
    x=mental_health_df["Work_Location"],
    y=mental_health_df["Number_of_Virtual_Meetings"],
    mode='markers'
    )
])


# Add dropdown
plot.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                ),
                dict(
                    args=["type", "box"],
                    label="Box Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "violin"],
                    label="Violin Plot",
                    method="restyle"
                )
            ]),
            direction="down",
        ),
    ]
)

plot.show()
```

# Heatmap

```python

# Select numerical columns for the correlation matrix
numerical_columns = [
    'Age', 'Years_of_Experience', 'Hours_Worked_Per_Week',
    'Number_of_Virtual_Meetings', 'Social_Isolation_Rating'
]

# Compute the correlation matrix for the numerical columns
corr_matrix = mental_health_df[numerical_columns].corr()
corr_matrix
```

```python

# Create the heatmap using Plotly Express
fig = px.imshow(corr_matrix,
                color_continuous_scale='Blues',  # Use the "Blues" color scale
                title="Correlation Heatmap of Numerical Variables",
                labels={'color': 'Correlation Coefficient'})  # Label for the color bar

# Show the figure
fig.show()
```

# Treemap

## Treemap of Employees by Region and Gender

```python
# Treemap of Employees by Region and Gender
fig = px.treemap(mental_health_df,
                 path=['Region', 'Gender'],  # Specify the hierarchy: Region > Gender
                 title='Treemap of Employees by Region and Gender',
                 color='Gender',  # Color by Gender
                 hover_data=['Region', 'Gender'], # Show Region and Gender on hover
                 color_discrete_map={
                     'Female': 'lightgreen',
                     'Male': 'skyblue',
                     'Prefer not to say': 'lightgray',
                     'Non-binary': 'lightyellow'

                 })

fig.show()
```

## Treemap Grouped by Job_Role and Stress_Level

```python
# Treemap Grouped by Job_Role and Stress_Level
fig = px.treemap(mental_health_df,
                 path=['Job_Role', 'Stress_Level'],  # Specify the hierarchy: Job Role > Stress Level
                 title='Treemap of Employees by Job Role and Stress Level',
                 color='Stress_Level',  # Color by Stress Level
                 hover_data=['Job_Role', 'Stress_Level'],  # Show Job Role and Stress Level on hover
                 color_discrete_map={
                     'Low': 'lightyellow',
                     'Medium': 'lightgreen',
                     'High': 'green',

                 })

fig.show()
```
